```julia
using Base: tan, tand, append!, eachindex, union, intersect, push!, deg2rad
using Core: ArgumentError, throw

export LED_selection

function LED_selection(user,user_num,led,led_num,height,device_height,ψ_c,degorrad)

    # grouping
    s_group = []
    user_group = []

    if degorrad = "deg"
        r = (height - device_height) * tan(ψ_c)
    elseif degorrad = "rad"
        r = (height - device_height) * tand(ψ_c)
    else 
        throw(ArgumentError("Wrong argument type degorrad"))
    end
    # LED mark
    for k in 1:user_num
        s_k = []
        for n in 1:led_num
            if (user[k,1] - led[n,1])^2 + (user[k,2] - led[n,2])^2 <= r^2
                s_k = append!(s_k,n)
                println(1)
            end
        end
        change_remark = 0
        for i in eachindex(s_group)
            if intersect(s_group[i],s_k) != []
                s_group[i] = union(s_group[i],s_k)
                user_group[i] = append!(user_group[i],k)
                change_remark = 1
                break    
            end
        end 
        if change_remark == 0
            s_group = push!(s_group,s_k)
            user_group = push!(user_group,k)
        end
    end

    return s_group, user_group
end

user = [6.0 6.0 0.85;
        34.0 6.0 0.85;
        34.0 34.0 0.85;
        6.0 34.0 0.85;
        20.0 10.0 0.85;
        20.0 30.0 0.85]
user_num = size(user,1)
led = [1.2 3.72 3.95;
    10.6 3.72 3.95;
    20.0 3.72 3.95;
    29.4 3.72 3.95;
    38.8 3.72 3.95;
    5.9 11.86 3.95;
    15.3 11.86 3.95;
    24.7 11.86 3.95;
    34.1 11.86 3.95;
    1.2 20.0 3.95;
    10.6 20.0 3.95;
    20.0 20.0 3.95;
    29.4 20.0 3.95;
    38.8 20.0 3.95;
    5.9 28.14 3.95;
    15.3 28.14 3.95;
    24.7 28.14 3.95;
    34.1 28.14 3.95;
    1.2 36.28 3.95;
    10.6 36.28 3.95;
    20.0 36.28 3.95;
    29.4 36.28 3.95;
    38.8 36.28 3.95]
led_num = size(led,1)
height = 3.95
device_height = 0.85

ψ_c = 60

s_group, user_group = LED_selection(user, user_num, led, led_num, height, device_height, ψ_c)

display(s_group)
display(user_group)
display(led[user_group[1],:])
```
