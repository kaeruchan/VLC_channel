```python
import math
from typing import List

class Solution:
  def minSubArrayLen(self, s: int, nums: List[int]) -> int:
    ans = math.inf
    sum = 0
    j = 0

    for i, num in enumerate(nums):
      sum += num
      while sum >= s:
        ans = min(ans, i - j + 1)
        sum -= nums[j]
        j += 1

    return ans if ans != math.inf else 0


print(Solution.minSubArrayLen(0,2,[1,3,4]))
```
